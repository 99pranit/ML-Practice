```python
import pandas as pd
import numpy as np
```

```python
credit_df = pd.read_csv('Credit_default_dataset.csv')
```

```python
credit_df.drop(columns= 'ID' , axis= 1, inplace= True)
```

```python
credit_df['MARRIAGE'] = credit_df['MARRIAGE'].map({0:3, 1:1, 2:2, 3:3})
credit_df['EDUCATION'] = credit_df["EDUCATION"].map({0:4,1:1,2:2,3:3,4:4,5:4,6:4})
```

```python
dummies_df = pd.get_dummies(credit_df, columns= ['MARRIAGE' , 'EDUCATION' , 'SEX'], drop_first= True)
```

```python
credit_df = pd.concat([credit_df , dummies_df] , axis= 1)
```

```python
credit_df.drop(columns=['MARRIAGE' , 'EDUCATION' , 'SEX'] , axis= 1)
```

```python
y = credit_df['default.payment.next.month']
x = credit_df.drop(columns= ['default.payment.next.month'])
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = scale.fit_transform(x)
```

```python
params={"n_estimators": [1,30,100,400,1000], 
"criterion" : ['gini', 'entropy'],
 "max_depth" : [ 3, 4, 8, 12, 15 , None]    
}
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
est = RandomForestClassifier()
rndsrh = RandomizedSearchCV(est , param_distributions= params , n_iter= 5, n_jobs= -1 , cv = 10)
```

```python
rndsrh.fit(x , y)
```

```python
rndsrh.best_params_
```

```python
from sklearn.model_selection import cross_val_score
```

```python
est = RandomForestClassifier(n_estimators= 30 , max_depth= 8 , criterion= 'entropy')
score = cross_val_score(est , x  , y ,cv= 10)
```
